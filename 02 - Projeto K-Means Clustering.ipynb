* Private: um fator com níveis Sim e Não indicando universidade privada ou pública

* Books: custos estimados dos livros
* Personal: gastos pessoais estimados
* PhD: porcentagem do corpo docente com doutorado
* Terminal: porcentagem do corpo docente com grau terminal
* S.F.Ratio: razão aluno/professor
* perc.alumni: porcentagem ex-alunos que doam
* Expend: despesas instrucionais por aluno
* Grad.Rate: taxa de graduação
